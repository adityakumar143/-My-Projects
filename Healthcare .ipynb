```python
hospital_counts = df['Hospital'].value_counts()

top_5_hospitals = hospital_counts.head(5)

plt.figure(figsize=(10, 6))
top_5_hospitals.plot(kind='bar', color='skyblue')
plt.title('The most top 5 hospitals')
plt.xlabel('hospitals')
plt.ylabel('count of patients')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

```python
# 1. What is the distribution of patient ages?


plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title('Distribution of Patient Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
bins = [0, 20, 40, 60, 80]
labels = ['0-20', '21-40', '41-60', '61-80']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_group_counts = df['Age Group'].value_counts()

top_5_age_groups = age_group_counts.head(5)

plt.figure(figsize=(10, 6))
top_5_age_groups.plot(kind='bar', color='lightgreen')
plt.title('top 5')
plt.xlabel('ages ')
plt.ylabel('count of patients')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title("Gender Distribution in Dataset")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#3. What are the most common medical conditions?

plt.figure(figsize=(12, 6))
top_conditions = df['Medical Condition'].value_counts().head(10)
sns.barplot(x=top_conditions.index, y=top_conditions.values, palette="viridis")
plt.title("Top 10 Most Common Medical Conditions")
plt.xlabel("Medical Condition")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python

 #4. How are patients distributed by blood type?
    
    
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Blood Type'], order=df['Blood Type'].value_counts().index, palette="coolwarm")
plt.title("Distribution of Blood Types")
plt.xlabel("Blood Type")
plt.ylabel("Count")
plt.show()
```

```python
# 5. What is the distribution of billing amounts?
#Distribution of billing amounts

plt.figure(figsize=(10, 5))
sns.histplot(df['Billing Amount'], bins=30, kde=True, color='red')
plt.title('Distribution of Billing Amounts')
plt.xlabel('Billing Amount ($)')
plt.ylabel('Count')
plt.show()
```

```python
# 6.Which hospitals have the highest number of admissions?

plt.figure(figsize=(12, 6))
top_hospitals = df['Hospital'].value_counts().head(10)
sns.barplot(x=top_hospitals.index, y=top_hospitals.values, palette="mako")
plt.title("Top 10 Hospitals with Most Admissions")
plt.xlabel("Hospital")
plt.ylabel("Number of Admissions")
plt.xticks(rotation=45)
plt.show()
```
